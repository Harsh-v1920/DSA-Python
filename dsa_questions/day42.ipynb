```python
#Leet code 
#Symmetric tree
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        def check(i,j):
            # if (i.left is None or i.right is None) and (j.left is None or j.right is None):
            #     return True
            if i is None and j is None:
                return True
            elif (i is None and j is not None) or (i is not None and j is None):
                return False
            if i.val==j.val:
                smallOutput1=check(i.left,j.right)
                smallOutput2=check(i.right,j.left)
                if smallOutput1 and smallOutput2 :
                    return True
                else:
                    return False                
            else:
                return False


        ans=check(root.left,root.right)
        return  ans
```

```python
#Longest substring without repeating characters
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        
        i,j=0,0
        lst=[]
        m=1
        n=len(s)
        flag=True
        if n==0:
            return 0
        while j<n:
            if s[j] not in lst:
                lst.append(s[j])
                j+=1
            else:
                flag=False
                m=max(m,len(lst))#ans
                lst.clear()
                i=i+1
                j=i
        # print(lst)
        if flag:
            return len(lst)
        else:
            m=max(m,len(lst))
            return(m)
                
```

```python
#Majority Element
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        d={}
        n=len(nums)
        m=0
        t=0
        if n==1:
            return 1
        for i in range(n):
            if nums[i] not in d:
                d[nums[i]]=1
            else:
                d[nums[i]]+=1
                if d[nums[i]]>n/2:
                    m=max(m,d[nums[i]])
                    if m==d[nums[i]]:
                        t=nums[i]
        return t
```
